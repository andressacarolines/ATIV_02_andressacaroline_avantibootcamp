## 1. Escreva uma função que receba uma lista de números e retorne outra lista com os números ímpares.


```python
def impares(lista):
    # Lista para armazenar os números ímpares
    impares = []

    # Percorre cada número na lista
    for numero in lista:
        if numero % 2 != 0:  # Verifica se o número é ímpar
            impares.append(numero)

    return impares


numeros = []
print("Digite os números (digite 'sair' para parar):")
while True:
    entrada = input("Número: ")

    if entrada.lower() == "sair":  # Se o usuário digitar 'sair', o loop para
        break

    # Converte a entrada para inteiro e adiciona à lista
    try:
        numeros.append(int(entrada))
    except ValueError:
        print("Por favor, digite um número válido ou 'sair' para terminar.")

resultado = impares(numeros)
print("Números ímpares:", resultado)
```

## 2. Escreva uma função que receba uma lista de números e retorne outra lista com os números primos presentes.

```python
# Função para verificar se um número é primo
def eh_primo(numero):
    if numero < 2:  # 0 e 1 não são primos
        return False
    for i in range(2, int(numero ** 0.5) + 1):  # Checa divisores até a raiz quadrada
        if numero % i == 0:
            return False
    return True

# Função para filtrar os números primos de uma lista
def filtrar_primos(lista):
    primos = []
    for num in lista:
        if eh_primo(num):
            primos.append(num)
    return primos

numeros = []  # Lista para armazenar os números digitados

print("Digite os números (digite 'sair' para encerrar):")
while True:
    entrada = input("Número: ")

    if entrada.lower() == "sair":
        break

    try:
        numeros.append(int(entrada))  # Converte para inteiro
    except ValueError:
        print("Por favor, digite um número válido ou 'sair' para encerrar.")

# Chamando a função
resultado = filtrar_primos(numeros)
print("Números primos encontrados:", resultado)
```

## 3. Escreva uma função que receba duas listas e retorne outra lista com os elementos que estão presentes em apenas uma das listas.


```python
# Função para pegar elementos exclusivos de duas listas
def elementos_exclusivos(lista1, lista2):

    exclusivos = []

    for elemento in lista1:
        if elemento not in lista2:  # Se está só na lista1
            exclusivos.append(elemento)

    for elemento in lista2:
        if elemento not in lista1:  # Se está só na lista2
            exclusivos.append(elemento)

    return exclusivos

def ler_lista(numeros):
    lista = []
    print(f"Digite os números da {numeros} lista (digite 'sair' para encerrar):")
    while True:
        entrada = input("Número: ")
        if entrada.lower() == "sair":
            break
        try:
            lista.append(int(entrada))
        except ValueError:
            print("Por favor, digite um número válido ou 'sair'.")
    return lista

# Lendo as duas listas
lista1 = ler_lista("primeira")
lista2 = ler_lista("segunda")

# Chamando a função
resultado = elementos_exclusivos(lista1, lista2)
print("Elementos exclusivos:", resultado)
```

## 4. Dada uma lista de números inteiros, escreva uma função para encontrar o segundo maior valor na lista.


```python
def segundo_maior(lista):
    # Primeiro, garantimos que a lista tem pelo menos 2 elementos
    if len(lista) < 2:
        return None  # Não dá para ter segundo maior

    # Remove duplicados para evitar problema, como [5, 5, 5]
    lista_unica = list(set(lista))

    # Ordena a lista em ordem decrescente
    lista_unica.sort(reverse=True)

    # O segundo maior será o segundo elemento
    return lista_unica[1]


numeros = []

print("Digite os números (digite 'sair' para encerrar):")
while True:
    entrada = input("Número: ")
    if entrada.lower() == "sair":
        break
    try:
        numeros.append(int(entrada))
    except ValueError:
        print("Por favor, digite um número válido ou 'sair'.")

resultado = segundo_maior(numeros)

if resultado is not None:
    print("O segundo maior número é:", resultado)
else:
    print("A lista não tem elementos suficientes para encontrar o segundo maior.")
```

## 5. Crie uma função que receba uma lista de tuplas, cada uma contendo o nome e a idade de uma pessoa, e retorne a lista ordenada pelo nome das pessoas em ordem alfabética.

```python
# Função para ordenar lista de pessoas pelo nome
def ordenar_por_nome(lista_pessoas):
    # key=lambda x: x[0] -> significa que vai ordenar pelo primeiro elemento da tupla (o nome)
    return sorted(lista_pessoas, key=lambda x: x[0])


pessoas = []

print("Digite os nomes e idades (digite 'sair' no nome para encerrar):")
while True:
    nome = input("Nome: ")
    if nome.lower() == "sair":
        break

    try:
        idade = int(input("Idade: "))
        pessoas.append((nome, idade))  # Cria uma tupla (nome, idade)
    except ValueError:
        print("Por favor, digite uma idade válida.")

resultado = ordenar_por_nome(pessoas)

print("\nLista ordenada pelo nome:")
for nome, idade in resultado:
    print(f"{nome} - {idade} anos")
```

## 6. Como identificar e tratar outliers em uma coluna numérica usando desvio padrão ou quartis?


```python
# Usando Desvio Padrão

import numpy as np
import pandas as pd

# Criando um dataset de exemplo
dados = pd.DataFrame({
    "idade": [18, 19, 20, 21, 22, 23, 50, 21, 22, 19, 20, 120]
})

# Calculando média e desvio padrão
media = dados["idade"].mean()
desvio = dados["idade"].std()

# Identificando outliers
limite_inferior = media - 3 * desvio
limite_superior = media + 3 * desvio

outliers = dados[(dados["idade"] < limite_inferior) | (dados["idade"] > limite_superior)]

print("Outliers encontrados:\n", outliers)

# Agora vou trocar por mediana

dados["idade"] = np.where(
    (dados["idade"] < limite_inferior) | (dados["idade"] > limite_superior),
    dados["idade"].median(),
    dados["idade"]
)

```

## 7. Como concatenar vários DataFrames (empilhando linhas ou colunas), mesmo que tenham colunas diferentes? Dica: Utiliza-se pd.concat() especificando axis=0 (linhas) ou axis=1 (colunas). Quando há colunas diferentes, os valores ausentes são preenchidos com NaN.

```python
import pandas as pd

# DataFrame 1
df1 = pd.DataFrame({
    "id": [1, 2, 3],
    "nome": ["Ana", "Bruno", "Carlos"]
})

# DataFrame 2 com uma coluna diferente
df2 = pd.DataFrame({
    "id": [4, 5],
    "idade": [25, 30]
})

# Concatenando pelas LINHAS
df_linhas = pd.concat([df1, df2], axis=0)

print(df_linhas)

```

```python
# DataFrame 3
df3 = pd.DataFrame({
    "cidade": ["São Paulo", "Rio de Janeiro", "Belo Horizonte"]
})

# Concatenando pelas COLUNAS
df_colunas = pd.concat([df_linhas, df3], axis=1)

print(df_colunas)
```

## 10.Utilizando pandas, como lidar com valores ausentes (NaN) em um DataFrame?

```python
import pandas as pd
import numpy as np

# Criando um DataFrame de exemplo
df = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", None, "Eduardo"],
    "idade": [25, np.nan, 22, 28, 35],
    "cidade": ["São Paulo", "Rio de Janeiro", None, "Salvador", "Fortaleza"]
})

print(df)
```

```python
# Verificar onde há NaN:
print("Verificar onde há NaN:")
print(df.isnull())

print("\n")

print("Contar valores ausentes por coluna:")

print(df.isnull().sum())
```

```python
# Agora para tratar esses valores, pode ser removido as linhas que tem valores ausêntes ou ser feita a imputação que seria substituir pela média ou mediana

df_sem_nan = df.dropna()
print(df_sem_nan)

```

```python
# Imputação (substiruir nomes por "desconhecido" e números pela media)

df["idade"] = df["idade"].fillna(df["idade"].mean())  # para números
df_preenchido = df.fillna("Desconhecido")  # para strings
df = df_preenchido

df["idade"] = df["idade"].fillna(df["idade"].mean())

df_preenchido
```
